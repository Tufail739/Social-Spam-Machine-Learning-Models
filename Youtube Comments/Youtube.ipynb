# 4. Data Cleaning

```python
#Converting class labels to binary values such as ham=0,Spam=1
from sklearn.preprocessing import LabelEncoder

encode=LabelEncoder()
f = encode.fit_transform(Y['Class'])
Y["Label"]=f
Y.head()
```

```python
#New column "Refined YT Comment" created to see changes
Y["Refined YT Comment"] = Y["YT_Comment"].copy()
Y.head()
```

```python
# Getting rid of all unwanted words,tabs..etc and a space is placed in all number places
Y["Refined YT Comment"] = Y["Refined YT Comment"].str.replace(r'\d+(\.\d+)?', '')
Y["Refined YT Comment"] = Y["Refined YT Comment"].str.replace(r'[^\w\d\s]', ' ')
Y["Refined YT Comment"] = Y["Refined YT Comment"].str.replace(r'\s+', ' ')
Y["Refined YT Comment"] = Y["Refined YT Comment"].str.replace(r'^\s+|\s+?$', '')

# Substituting URL addresses with 'add'
Y["Refined YT Comment"] = Y["Refined YT Comment"].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$',
                                  'add')

# Substituting currency symbols with 'currency'
Y["Refined YT Comment"] = Y["Refined YT Comment"].str.replace(r'£|\$', 'currency')

# Changing all capital words to small word
Y["Refined YT Comment"] = Y["Refined YT Comment"].str.lower()

# Getting rid of 1 term words
Y["Refined YT Comment"] = Y["Refined YT Comment"].str.replace(r'\b\w\b', '').str.replace(r'\s+', ' ')
Y.head()
```

```python
# Removing punctuation marks like . , ! , ? , : ..etc
s.punctuation
```

# 5. Data Preprocessing

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer as ss
from nltk.corpus import stopwords
#gives a specific set of words 
Sw= (stopwords.words('english')) 
Sw[0:10] 
```

```python
#Removing terms using the stemming process
processStem = ss("english")
def txtproc(cmt):   
    cmt = cmt.translate(str.maketrans('','', s.punctuation))
    ws = [processStem.stem(word) for word in cmt.split() if word.lower() not in stopwords.words("english")]
    return " ".join(ws)
Y["Refined YT Comment"] = Y["Refined YT Comment"].apply(txtproc)
```

```python
#showing the old YT Comment and new YT Comments transformed
print("Unprocessed 2nd YT Comment:", Y['YT_Comment'][2])
print("\nRefined YT Comment :", Y["Refined YT Comment"][2], '\n')

print("Unprocessed 8th YT Comment:", Y['YT_Comment'][8])
print("\nRefined YT Comment :", Y["Refined YT Comment"][8], '\n')

print("Unprocessed 300th YT Comment:", Y['YT_Comment'][300])
print("\nRefined YT Comment :", Y["Refined YT Comment"][300], '\n')

print("Unprocessed 17th YT Comment:", Y['YT_Comment'][17])
print("\nRefined YT Comment :", Y["Refined YT Comment"][17], '\n')

print("Unprocessed 17th YT Comment:", Y['YT_Comment'][239])
print("\nRefined YT Comment :", Y["Refined YT Comment"][239], '\n')
```

```python
#token are created with YT Comments refined
from nltk import word_tokenize,sent_tokenize
Y['Toklist'] = Y['Refined YT Comment'].map(lambda w: nltk.tokenize.word_tokenize(w))
```

# 6.Splitting the dataset into training and testing data 

```python
from sklearn import feature_extraction, model_selection, naive_bayes, svm
from sklearn import metrics as m
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix as CT
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import classification_report as ClassRep
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import learning_curve
from sklearn.utils import resample
from sklearn.metrics import roc_curve
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

TrainA, TestA, TrainB, TestB = train_test_split(Wholedataset['YT_Comment'],
                                                    Wholedataset['Label'],
                                                    test_size=0.3,
                                                    random_state=101)
```

## 6a. Balancing the Dataset

```python
# Joining TrainA and TrainB column to create a whole dataset consisting of Content and label columns
TrainANewset=pan.concat([TrainA, TrainB], axis=1)

NonSpam=TrainANewset[TrainANewset['Label']==0] 

Spam=TrainANewset[TrainANewset['Label']==1] 

#Upgrading the dataset
SpamUpgrade = resample(Spam,replace=True,n_samples=680,random_state=101) 

# joining upgraded spam messages with non-spam messages for a unified distribution
Upgraded = pan.concat([NonSpam, SpamUpgrade])

TrainA=Upgraded['YT_Comment']
TrainB=Upgraded['Label']
```

```python
#Visualizing the new balanced dataset

Upgraded["Label"].value_counts().plot(kind = 'pie', 
                                          figsize = (6, 7))
                                          

pt.title('Spam and Non-Spam Evenly Distributed')
pt.legend(["Non-spam YT terms", "Spam YT terms"])
pt.show()
```

## 6b.Vectorization process using TF-IDF Vectorizer

```python
# Train data is transformed using TFIDF 
Tfvecto = TfidfVectorizer(stop_words = 'english', use_idf = True, lowercase = True, strip_accents='ascii')
TrainA_Tfvecto = Tfvecto.fit_transform(TrainA) 
TestA_Tfvecto = Tfvecto.transform(TestA) 
print(TrainA_Tfvecto.shape)
```

## 7.1 Naive Bayes

```python
# Model training and prediction
NBModel = MultinomialNB()
NBModel.fit(TrainA_Tfvecto, TrainB)
```

```python
NBPred = NBModel.predict(TestA_Tfvecto)
accuracy1=m.accuracy_score(TestB, NBPred)*100
print(accuracy1)
```

```python
print(ClassRep(TestB, NBPred))

pan.DataFrame(
   CT(TestB, NBPred),
    index = [['ActualData', 'ActualData',], ['Ham', 'Spam']],
    columns = [['PredictedData', 'PredictedData'], ['Ham', 'Spam']])
```

```python

fpr1, tpr1, thresholds1 = m.roc_curve(TestB, NBPred, pos_label=1)
NB_Curve = m.auc(fpr1, tpr1)
fig1=pt.figure(figsize=(6, 4), dpi= 80, facecolor='w', edgecolor='k')

print("Naive Bayes Area Under Curve is : {0}".format(NB_Curve))

pt.title('Naive Bayes ROC')
pt.plot(fpr1, tpr1,linestyle='--', label='Naive Bayes AUC: %0.3f' %NB_Curve)
pt.legend(loc = 'lower right')
pt.plot([0, 1], [0, 1],'r--')
pt.xlim([0, 1])
pt.ylim([0, 1])
pt.ylabel('True +ve Rate')
pt.xlabel('False +ve Rate')
pt.savefig('NB ROC')
pt.grid()
pt.show()
```
